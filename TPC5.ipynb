### TPC5: Análise de dados: doença cardíaca

Descarregue o ficheiro de dados: `myheart.csv`
Crie um programa em Python, conjunto de funções, que responda às seguintes questões:
* Crie uma função que lê a informação do ficheiro para um modelo, previamente pensado em memória;
* Crie uma função que calcula a distribuição da doença por sexo;
* Crie uma função que calcula a distribuição da doença por escalões etários. Considere os seguintes escalões: [30-34], [35-39], [40-44], ...
* Crie uma função que calcula a distribuição da doença por níveis de colesterol. Considere um nível igual a um intervalo de 10 unidades, comece no limite inferior e crie os níveis necessários até abranger o limite superior;
* Crie uma função que imprime na forma de uma tabela uma distribuição;
* Especifique um programa que ao executar apresenta as tabelas correspondentes às distribuições pedidas.

dataset: 
    Age,Sex,ChestPainType,RestingBP,Cholesterol,FastingBS,RestingECG,MaxHR,ExerciseAngina,Oldpeak,ST_Slope,HeartDisease

```python
def getPaciente(texto):
    paciente=[]
    campos=texto.split(',')
    paciente.append(campos[0])
    paciente.append(campos[1])
    paciente.append(campos[3])
    paciente.append(campos[5])
    return paciente
    


def lerDataset(fnome):
    f=open(fnome, encoding ="utf-8")
    bd=[]
    f.readline()
    for linha in f:
        bd.append(getPaciente(linha))
    return bd
```

```python
BD=[]
BD=lerDataset('../ATP/myheart.csv')
print(BD)
```

```python
def chaveOrd(a):
    return a[2]
    

def listarDataset(bd):
    bd.sort(key=chaveOrd)
    print('id      sexo      colesterol      doença')
    print('________________________________________')
    for a in bd:
        print(a[0] + "    |     "+a[1]+"     |     "+a[2]+"     |   + "+a[3])

listarDataset(BD)
```

```python
def distribPorSexo(d):
    distribuicao={}
    for a in d:
        if a[1] in distribuicao.keys():
            distribuicao[a[1]]=distribuicao[a[1]]+1
        else:
            distribuicao[a[1]]=1
    return distribuicao
    
d=distribPorSexo(BD)
print(d)
```

```python
def distribPorIdade(d):
    distribuicao={}
    novodic={'[30-34]':0,'[35-39]':0,'[40-44]':0,'[45-49]':0,'[50-54]':0,'[55-59]':0,'[60-64]':0,'[65-69]':0,'[70-74]':0,'[75-79]':0}
    for a in d:
        if a[0] in distribuicao.keys():
            distribuicao[a[0]]=distribuicao[a[0]]+1
        else:
            distribuicao[a[0]]=1
    for key in distribuicao.keys():
        if  30 <= int(key) <= 34:
            novodic['[30-34]']=novodic['[30-34]']+distribuicao[key]
        
        elif  30 <= int(key) <= 34:
            novodic['[30-34]']=novodic['[30-34]']+distribuicao[key]
        elif  35 <= int(key) <= 39:
            novodic['[35-39]']=novodic['[35-39]']+distribuicao[key]
        elif  40 <= int(key) <= 44:
            novodic['[40-44]']=novodic['[40-44]']+distribuicao[key]
        elif  45 <= int(key) <= 49:
            novodic['[45-49]']=novodic['[45-49]']+distribuicao[key]
        elif  50 <= int(key) <= 54:
            novodic['[50-54]']=novodic['[50-54]']+distribuicao[key]
        elif  55 <= int(key) <= 59:
            novodic['[55-59]']=novodic['[55-59]']+distribuicao[key]
        elif  60 <= int(key) <= 64:
            novodic['[60-64]']=novodic['[60-64]']+distribuicao[key]
        elif  65 <= int(key) <= 69:
            novodic['[65-69]']=novodic['[65-69]']+distribuicao[key]
        elif  70 <= int(key) <= 74:
            novodic['[70-74]']=novodic['[70-74]']+distribuicao[key]
        elif  75 <= int(key) <= 79:
            novodic['[75-79]']=novodic['[75-79]']+distribuicao[key]
    return novodic
  
d=distribPorIdade(BD)
print(d)
```

```python
def distribPorColes(d):
    distribuicao1={}
    novo={'[100-110]':0,'[110-120]':0,'[120-130]':0,'[130-140]':0,'[140-150]':0,'[150-160]':0,'[160-170]':0,'[170-180]':0,'[180-190]':0,'[190-200]':0,
         }
    for a in d:
        if a[2] in distribuicao1.keys():
            distribuicao1[a[2]]=distribuicao1[a[2]]+1
        else:
            distribuicao1[a[2]]=1
        for key in distribuicao1.keys():
            if  100 <= int(key) < 110:
                novo['[100-110]']=novo['[100-110]']+distribuicao1[key]
            elif 110 <= int(key) < 120:
                novo['[110-120]']=novo['[110-120]']+distribuicao1[key]
            elif  120 <= int(key) < 130:
                novo['[120-130]']=novo['[120-130]']+distribuicao1[key]
            elif  130 <= int(key) < 140:
                novo['[130-140]']=novo['[130-140]']+distribuicao1[key]
            elif  140 <= int(key) < 150:
                novo['[140-150]']=novo['[140-150]']+distribuicao1[key]
            elif  150 <= int(key) < 160:
                novo['[150-160]']=novo['[150-160]']+distribuicao1[key]
            elif  160 <= int(key) < 170:
                novo['[160-170]']=novo['[160-170]']+distribuicao1[key]
            elif  170 <= int(key) < 180:
                novo['[170-180]']=novo['[170-180]']+distribuicao1[key]
            elif  180 <= int(key) < 190:
                novo['[180-190]']=novo['[180-190]']+distribuicao1[key]
            elif  190 <= int(key) < 200:
                novo['[190-200]']=novo['[190-200]']+distribuicao1[key]
        
    return novo
    
d=distribPorColes(BD)
print(d)
```

```python
import matplotlib.pyplot as plt
def plotDistribPorSexo(d):
    a=distribPorSexo(d)
    tuplo=a.items()
    x=[]
    y=[]
    for elem in tuplo:
        x.append(elem[0])
    for elem in tuplo:
        y.append(elem[1])
    plt.bar(x,y,color=['green','black'])
    plt.xlabel('Sexo')
    plt.ylabel('Número de pessoas')
    barWidth=0.25
    plt.title('Distribuição de pacientes por sexo')
    plt.show()

plotDistribPorSexo(BD)
```
